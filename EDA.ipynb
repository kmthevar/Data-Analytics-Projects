# Columns with null values

```python
#finding all the columns containing null values and adding their names to a new list
na_cols=[]
for cols in dataset.columns:
    print(cols,' = ',dataset[cols].isnull().sum())
    if dataset[cols].isnull().sum()>1:
        na_cols.append(cols)
        
```

```python
na_cols
```

```python
#calculating the % of missing values--isnull().mean() will calculate mean only for null values wrt to all the values
print("% of missing values in the columns")
for cols in na_cols:
    print(cols,'=',np.round(dataset[cols].isnull().mean()*100,4))
```

```python
#Assigning 1 to null and 0 to non null values inside the respective columns

for cols in na_cols:
    data=dataset.copy()
   
    #Assigning 1 to null and 0 to non null values inside the respective columns
    data[cols]=np.where(data[cols].isnull(), 1, 0)
    
    #Calculating the median of house prices for null and non null values for features containing null
    data.groupby(cols)['SalePrice'].median().plot.bar(color=('b','r'),rot=0)
    plt.title(cols)
    plt.show()
    
```

From the above graphs we can see there is a relation between sales price(dependent variable) and features with NaN.
We need to handle these values in Feature Engineering

# Numerical Variables

```python
#Getting all the columns containing numerical values
cols_num=[]
for cols in dataset.columns:
    if dataset[cols].dtype != object:
        cols_num.append(cols)
        
```

```python
dataset[cols_num]
```

# Temporal Variables (eg: datetime variables)

```python
# Getting the columns which contain date values
cols_year=[]
for cols in cols_num:
    if 'Yr' in cols or 'Year' in cols:
        cols_year.append(cols)
```

```python
cols_year
```

```python
for cols in cols_year:
    print(cols,dataset[cols].unique())
```

```python
#Finding relation between year sold and sales price

dataset.groupby('YrSold')['SalePrice'].median().plot(xticks=dataset['YrSold'].unique())
plt.xlabel('Year Sold')
plt.ylabel('SalePrice')
plt.title('House Price vs Year Sold')
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot.bar()
plt.title('House Price vs Year Sold')
```

```python
#Relation between Sales Price and features like Year Built, Remodel date and graage built date
for cols in cols_year:
    if cols !='YrSold':
        plt.scatter(dataset[cols],dataset['SalePrice'])
        plt.xlabel(cols)
        plt.ylabel('Sales Price')
        plt.show()
```

```python
#Relation between Saleprice and the number of years from features
#like Year Built, Remodel date and graage built wrt Year Sold (i.e age of the house wrt to selling date)
for cols in cols_year:
    if cols !='YrSold':
        data=dataset.copy()
        #below code will claculate the number of years between sold year and the respective year feature
        data[cols]=data['YrSold']-data[cols]
        plt.scatter(data[cols],data['SalePrice'])
        plt.xlabel(cols)
        plt.ylabel('Sales Price')
        plt.title(cols)
        plt.show()
```

# Continous variables

```python
#Finding continous variable from dataset

cols_cont=[cols for cols in cols_num if cols not in cols_dis+cols_year+['Id']]
print('Total number of dicrete columns =',len(cols_cont))
print('Discrete columns:\n',cols_cont)
```

```python
dataset[cols_cont].head()
```

```python
#Understanding the distribution of continous values
for cols in cols_cont:
    dataset[cols].hist(bins=25)
    plt.xlabel(cols)
    plt.ylabel('count')
    plt.title(cols)
    plt.show()
```

# Performing log transformation to remove skewness in data

```python
log_data=dataset.copy()
for cols in cols_cont:
    data=dataset.copy()
    if 0 in data[cols].unique():
        #print(cols)
        pass
    else:
        data[cols]=np.log(data[cols])
        log_data[cols]=np.log(log_data[cols])
        #print(dataset[cols],data[cols])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[cols],data['SalePrice'])
        plt.xlabel(cols)
        plt.ylabel('Sale Price')
        plt.title(cols)
        plt.show()
```

# Histogram of contnuous values after log transformation

```python

for cols in cols_cont:
    if 0 in dataset[cols].unique():
        #print(cols)
        pass
    else:
        log_data[cols].hist(bins=25)
        plt.xlabel(cols)
        plt.ylabel('count')
        plt.title(cols)
        plt.show()
```

# Outlier detection

```python

#works only for continous variables
for cols in cols_cont:
    data=dataset.copy()
    if 0 in data[cols].unique():
        #print(cols)
        pass
    else:
        data[cols]=np.log(data[cols])
        data.boxplot(column=cols)
        plt.ylabel(cols)
        plt.title(cols)
        plt.show()
```
